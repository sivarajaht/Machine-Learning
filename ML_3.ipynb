### Case Study II - Cereal Data

```python
cer1 = pd.read_csv('cereal.csv')
print(cer1.head())
cer1.info()
```

```python
print(cer1['sugars'].max(axis = 0))
print(cer1['sugars'].min())
```

```python
print(cer1['vitamins'].max())
print(cer1['vitamins'].min())
```

```python
vit = cer1['vitamins'].values
plt.hist(vit, 20, label='Vitamins Content')
plt.legend()
plt.title("Vitamins Content histogram")
plt.show()
```

```python
my_dict = {'mfr_name':{'N':'Nabisco',
                  'Q':'Quaker Oats',
                  'K':'Kelloggs',
                  'R':'Raslston Purina',
                  'G':'General Mills' ,
                  'P':'Post' ,
                  'A':'American Home Foods Products'
                 }
          }
```

```python
cer1['mfr_name'] = cer1['mfr']
cer1.replace(my_dict, inplace=True)
#cer1['mfr_name'] = cer1['mfr']
#for i in cer1.index():
#    cer1.iloc([i,'mfr_name']) = my_dict[cer1.iloc(['mfr_name'])]
#print(cer1['mfr_name'])
print(cer1)
```

```python
import seaborn as sns
ax = sns.countplot(y="mfr_name", data=cer1)
```

```python
y = cer1['rating'].values
```

```python
x = cer1.select_dtypes(include=['int64', 'float64']).copy()
x.drop('rating', axis=1, inplace=True)
print(x)
```

```python
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
"X_train = {}, X_test = {}, y_train = {}, y_test = {}".format(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
print(X_train)
```

```python
# fit a model
lm = linear_model.LinearRegression()
model = lm.fit(X_train, y_train)
predictions = lm.predict(X_test)
```

```python
print(y_test)
predictions[0:5]
```

```python
## The line / model
plt.scatter(y_test, predictions)
#plt.xlabel(â€œTrue Valuesâ€)
#plt.ylabel(â€œPredictionsâ€)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error
rms = np.sqrt(mean_squared_error(y_test, predictions))
print(rms)
```
